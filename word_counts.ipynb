```python
f = open("input.txt")
text = f.read()
f.close()
text = text.replace(" ","_")
text
```

```python
one_dict = {}
for i in range(len(text)):
    p = text[i]
    #print(pr)
    if p in one_dict:
        one_dict[p] += 1
    else:
        one_dict[p] = 1
dict(sorted(one_dict.items(), key=lambda item: item[1]))
```

```python
pair_dict = {}
for i in range(len(text)):
    if(i < len(text)-1):
        if text[i] < text[i+1]:
            pr = text[i]+text[i+1]
        else:
            pr = text[i+1]+text[i]
        #print(pr)
        if pr in pair_dict:
            pair_dict[pr] += 1
        else:
            pair_dict[pr] = 1
dict(sorted(pair_dict.items(), key=lambda item: item[1]))
```
